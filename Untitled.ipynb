```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('mushrooms.csv') #Le o dataset
display(df.head()) #mostra as 5 primeiras amostras
colum_names = df.columns
```

```python
for name in colum_names: #verifica o balanceamento do dataset
    print('============================================')
    print(df[name].value_counts())
```

```python
print("Dados faltantes em cada feature:")
print(df.isnull().sum())
```

```python
df = df.drop('veil-type', axis=1)   
print("\nDimensÃµes do dataset apÃ³s a remoÃ§Ã£o:", df.shape)

# A coluna 'class' Ã© o nosso alvo (label)
# p = poisonous (venenoso), e = edible (comestÃ­vel)
y = df['class']

# X sÃ£o todas as outras colunas (features)
X = df.drop('class', axis=1)

print("Label 'y' separado:")
print(y.head())
print("\nFeatures 'X' separadas:")
display(X.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Instanciar o codificador
le = LabelEncoder()

# Aplicar a codificaÃ§Ã£o em y
y_encoded = le.fit_transform(y)

# Mostrar a correspondÃªncia da codificaÃ§Ã£o
# (ex: 'e' -> 0, 'p' -> 1)
print(f"Labels originais: {le.classes_}")
print(f"Labels codificados: {np.unique(y_encoded)}")
print("\nExemplo dos 5 primeiros labels codificados:", y_encoded[:5])
```
